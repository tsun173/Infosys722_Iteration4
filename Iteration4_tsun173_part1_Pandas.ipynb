```python
#3.2. Clean the Data
import numpy as np
def getOutlierRates(dataset,colName):
    out_col = dataset[colName]
    out_collist = []
    for value in out_col:
        out_collist += [value]
    out_collist.sort()
    col_q1,col_q3 = np.percentile(out_collist,[25,75])
    
    col_iqr = col_q3 - col_q1
    lowerBound = col_q1 - (1.5 * col_iqr)
    upperBound = col_q3 + (1.5 * col_iqr)
    
    outliers = []
    for value in out_col:
        if value < lowerBound or value > upperBound:
            outliers += [value]
    return len(outliers)/len(dataset[colName])

print('2014_Coastal Outliers Rate: ',getOutlierRates(dataset_2014,'Coastal'))
print('2014_Blue Outliers Rate: ',getOutlierRates(dataset_2014,'Blue'))
print('2014_Green Outliers Rate: ',getOutlierRates(dataset_2014,'Green'))
print('2014_Red Outliers Rate: ',getOutlierRates(dataset_2014,'Red'))
print('2014_NIR Outliers Rate: ',getOutlierRates(dataset_2014,'NIR'))
print('2014_SWIR1 Outliers Rate: ',getOutlierRates(dataset_2014,'SWIR1'))
print('2014_SWIR2 Outliers Rate: ',getOutlierRates(dataset_2014,'SWIR2'))
print('2018_Coastal Outliers Rate: ',getOutlierRates(dataset_2018,'Coastal'))
print('2018_Blue Outliers Rate: ',getOutlierRates(dataset_2018,'Blue'))
print('2018_Green Outliers Rate: ',getOutlierRates(dataset_2018,'Green'))
print('2018_Red Outliers Rate: ',getOutlierRates(dataset_2018,'Red'))
print('2018_NIR Outliers Rate: ',getOutlierRates(dataset_2018,'NIR'))
print('2018_SWIR1 Outliers Rate: ',getOutlierRates(dataset_2018,'SWIR1'))
print('2018_SWIR2 Outliers Rate: ',getOutlierRates(dataset_2018,'SWIR2'))
```

```python
#3.3. Construct the data
from osgeo import gdal
from tqdm import tqdm
import numpy as np
import csv


with open('./soilCSV_2014_500000.csv','w',newline='') as writeCSVFile:
     writer = csv.writer(writeCSVFile)
     #writer.writerow(["Coastal","Blue","Green","Red","NIR","SWIR1","SWIR2","RVI","NDVI","NDWI","NDMI","Soil_Class"])
     writer.writerow(["Coastal", "Blue", "Green", "Red", "NIR", "SWIR1", "SWIR2","Soil_Class"])

     LabelingRaster = gdal.Open('./MSsoil_Label_NoDataToZero.tif')
     trainingRaster = gdal.Open("./2014-MS.tif")

     label = LabelingRaster.GetRasterBand(1).ReadAsArray()
     Coastal = trainingRaster.GetRasterBand(1).ReadAsArray()
     Blue = trainingRaster.GetRasterBand(2).ReadAsArray()
     Green = trainingRaster.GetRasterBand(3).ReadAsArray()
     Red = trainingRaster.GetRasterBand(4).ReadAsArray()
     NIR = trainingRaster.GetRasterBand(5).ReadAsArray()
     SWIR1 = trainingRaster.GetRasterBand(6).ReadAsArray()
     SWIR2 = trainingRaster.GetRasterBand(7).ReadAsArray()

     for i in tqdm(range(len(label))):
        for j in range(len(label[0])):
            if label[i][j] != 0 and Coastal[i][j] != 65535:
                coastalValue = float(Coastal[i][j])
                blueValue = float(Blue[i][j])
                greenValue = float(Green[i][j])
                redValue = float(Red[i][j])
                nirValue = float(NIR[i][j])
                swir1Value = float(SWIR1[i][j])
                swri2Value = float(SWIR2[i][j])
                RVI = nirValue / redValue
                NDVI = (nirValue-redValue)/(nirValue+redValue)
                NDWI = (greenValue-nirValue) / (greenValue+nirValue)
                NDMI = (nirValue-swir1Value) / (nirValue+swir1Value)
                labelValue = label[i][j]

                writer.writerow
                ([coastalValue,blueValue,greenValue,redValue,nirValue,swir1Value,swri2Value,RVI,NDVI,NDWI,NDMI,labelValue])
               # writer.writerow([coastalValue, blueValue, greenValue, redValue, nirValue, swir1Value, swri2Value,labelValue])
```
